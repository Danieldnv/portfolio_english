```python
import matplotlib.pyplot as plt
import yfinance as yf
import numpy as np
import pandas as pd
import seaborn as sns
```

```python

#Collecting data on both apple and tesla stocks
data = yf.download('AAPL', start= '2019-01-01', end= '2024-10-01')
data_tsla = yf.download('TSLA', start='2019-01-01', end='2024-10-01')
```

```python


'''the plt.style.use() must be before the plt.figure'''
#Some themes I tried:
#plt.style.use('fivethirtyeight')
#plt.style.use('ggplot')
#plt.style.use('Solarize_Light2')
#plt.style.use('classic') 
#plt.style.use('bmh')
#plt.style.use('dark_background') 
#plt.style.use('Tufte') 
#plt.style.use('tableau-colorblind10')

################plt.style.use('default') to go back to the original style


plt.figure(figsize=(12,4),dpi=300, facecolor='gainsboro')
#Plotting the closing prices for apple
plt.plot(data.index, data['Close'], label='AAPL Closing Prices', color='blue')
#Plotting the closing prices for tesla
plt.plot(data_tsla.index, data_tsla['Close'], label='TSLA Closing Prices', color='cornflowerblue')

#Discovering the price of apple stock during the pandemic crash
'''close_aapl= data.loc['2020-03-16']['Close']'''
'''print(close_aapl)'''

#Adding the arrow and the annotation to the pandemic crash price of apple.
plt.annotate('Pandemic', fontsize=8, xy=(pd.to_datetime('2020-03-16'),60.55), xytext=(pd.to_datetime('2020-03-16'),60.55 + 110),
             arrowprops=dict(facecolor='black', arrowstyle='->'))


#max prices for apple and tesla

# Identifying the maximum price of the apple stocks
max_aapl=data['Close'].max() 
max_tsla=data_tsla['Close'].max()

# Finding the correspondent date for the maximum price of apple
date_max_aapl=data['Close'].idxmax() #Idmax finds the index of the first highest occurence in a df. 
date_max_tsla=data_tsla['Close'].idxmax() # In this code, the index is already set to the 'Date' column because yfinance automatically does it
                      
#plotting the highest prices scatters (points, little dots)

plt.scatter(date_max_aapl,max_aapl, color='red', label= 'AAPL highest price', zorder=5, s=5)

plt.scatter(date_max_tsla,max_tsla, color='maroon', label='TSLA highest price', zorder=5, s=5)

#Aesthetics
plt.axvline(x=pd.to_datetime('2020-03-16'), color='dimgrey', linestyle=':', label='Pandemic crash',linewidth=1)
plt.title('Closing Prices of Tesla and Apple Stocks', fontsize=18, fontname='Times New Roman')
plt.xlabel('Date', labelpad=15, fontsize=16, fontname='Times New Roman')
plt.ylabel('Price (USD)', fontsize=15, labelpad = 15, fontname='Times New Roman')
plt.legend(fontsize=7)
plt.grid(visible=True, linestyle= '--', alpha=0.7)
plt.xticks(rotation=45, fontsize=12, fontname='Times New Roman')
plt.yticks( fontname='Times New Roman', fontsize=10)
plt.figtext(0.8, -0.1, "Source: Yahoo Finance", fontsize=10, color='gray')

plt.show()

```
